```python
import matplotlib.pyplot as plt
import numpy as np

def plot_loss(MODEL_NAME, TRAIN_TEST_RATIO):
    # Load training and test loss data
    train_loss_data = np.load(f'{MODEL_NAME}_train_loss.npy')
    test_loss_data = np.load(f'{MODEL_NAME}_test_loss.npy')
    train_loss_data = [np.mean(train_loss_data[i:i+TRAIN_TEST_RATIO]) for i in range(0,len(train_loss_data)-TRAIN_TEST_RATIO+1,TRAIN_TEST_RATIO)]
    test_train_diff = abs(test_loss_data-train_loss_data)
    num_steps = len(test_loss_data)

    def print_data_stat(title, data):
        print(f'\n{title}:')
        print('min:', min(data))
        print('mean:', np.mean(data))
        print('mean of last 50%:', np.mean(data[int(len(data) * 0.5):]))
        print('mean of last 10%:', np.mean(data[-int(0.1 * len(data)):]))

    # print some stat
    print('train-test ratio:', TRAIN_TEST_RATIO)
    print('number of steps:', num_steps)
    print_data_stat('Train Loss', train_loss_data)
    print_data_stat('Test Loss', test_loss_data)
    print_data_stat('Test Train Difference', test_train_diff)

    # plot data
    plt.plot(np.arange(len(train_loss_data)), train_loss_data, label='train loss', color='blue')
    plt.plot(np.arange(len(test_loss_data)), test_loss_data, label='test loss', color='red')
    plt.plot(np.arange(len(test_train_diff)), test_train_diff, label='test-train difference', color='green')

    plt.xlabel('step')
    plt.ylabel('loss')
    plt.legend()
    plt.title(MODEL_NAME)
    plt.show()
```

```python
plot_loss('minerl_navigator_5_epoch_1', 16)
plot_loss('minerl_navigator_5_epoch_2', 16)
plot_loss('minerl_navigator_5_epoch_3', 16)
```
